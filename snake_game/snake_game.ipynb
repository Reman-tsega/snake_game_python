```python
from turtle import Turtle
from turtle import Screen
# from snake_game.ipynb import Snake
# from snake_game.ipynb import Screen
import time
```

```python
# from turtle import Turtle, Screen
import random
STARTING_POSIRON = [(0,0), (-20,0),(-40,0)]
MOVED_DISTANCE =20





class Snake:
    snake_position=[]
    def __init__(self):
        self.body=[]
        self.create_snake()
        self.head = self.body[0]
        

    def create_snake(self):
        for x in range(3):
            tim = Turtle("turtle") 
            tim.color("white")
            tim.penup()
        #     tim.shapesize(2,2)
            tim.goto(STARTING_POSIRON[x])
            self.body.append(tim)
        
    def move_snake(self):
        for seg_num in range(len(self.body)-1,0,-1):
            x_next_seg = self.body[seg_num-1].xcor()
            y_next_seg = self.body[seg_num-1].ycor()
            self.body[seg_num].goto(x_next_seg, y_next_seg)
            Snake.snake_position.append( self.body[seg_num].xcor() )
        self.body[0].fd(MOVED_DISTANCE)
   

        
#___________________control the snack direction ______________________________-

    def up(self):
        turtle_heading = self.body[0].heading()
        if turtle_heading==0.0:
            self.body[0].left(90)
        elif turtle_heading==180.0:
            self.body[0].right(90)
    def down(self):
        turtle_heading = self.body[0].heading()
        if turtle_heading==0.0:
            self.body[0].right(90)
        elif turtle_heading==180.0:
            self.body[0].left(90)
    def left(self):
        turtle_heading = self.body[0].heading()
        if turtle_heading==90.0:
            self.body[0].left(90)
        elif turtle_heading==270.0:
            self.body[0].right(90)
    def right(self):
        turtle_heading = self.body[0].heading()
        if turtle_heading==90.0:
            self.body[0].right(90)
        elif turtle_heading==270.0:
            self.body[0].left(90)

```
